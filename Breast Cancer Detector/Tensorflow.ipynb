```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from tensorflow import keras
```

```python
from sklearn import datasets
```

```python
data = datasets.load_breast_cancer()
```

```python
data.keys()
```

```python
X = (data.data)
y = data.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=69)
```

```python
X_train.shape
```

```python
y_train.reshape((398,1))
```

```python
model = keras.Sequential()
```

```python
model.add(keras.layers.Flatten(input_shape=(1,30)))
model.add(keras.layers.Dense(32, activation='relu'))
model.add(keras.layers.Dense(16, activation='relu'))
model.add(keras.layers.Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
fitModel = model.fit(X_train, y_train, epochs=498)
```

```python
from sklearn.metrics import confusion_matrix, precision_score, precision_recall_curve
```

```python
predictions = model.predict(X_test)
predictions=predictions > 0.9
confusion_matrix(predictions, y_test)
```
